```python
import requests
from bs4 import BeautifulSoup as bs
from bs4.element import Tag
import re
```

```python
resp = requests.get('http://www.rxassist.org/pap-info/brand-drug-list-print')
page = bs(resp.text)
```

```python
names = []
rows = page.find_all('tr')
for row in rows:
    cells = row.find_all('td')
    if len(cells) != 2: continue
    names.append(cells[0].get_text())
names
```

```python
names_lower = [name.lower() for name in names]
```

```python
name_re = re.compile('(.*?)\s[^a-zA-Z]')
names_cleaned = []
for name in names_lower:
    matches = name_re.findall(name)
    if matches:
        for match in matches:
            names_cleaned.append(match)
    else:
        names_cleaned.append(name)
names_cleaned = list(set(names_cleaned))
```

```python
with open('drug_names.txt', 'w') as f:
    print('\n'.join(names_cleaned), file=f)
```

```python
window_size = 5

names_long = list(filter(lambda x: len(x) >= 5, names_cleaned))
name_windows = []
for name in names_long:
    windows = []
    for i in range (len(name) - window_size + 1):
        windows.append(name[i:i+window_size])
    name_windows.append(windows)
name_windows
```
